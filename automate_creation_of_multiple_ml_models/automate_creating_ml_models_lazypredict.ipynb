```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from IPython.core.display import display
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier,ExtraTreesClassifier
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor,ExtraTreesRegressor

from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score, r2_score
from sklearn.model_selection import train_test_split

# import lazypredict

from lazypredict.Supervised import LazyClassifier, LazyRegressor
```

## Classification Problem : LazyClassifier

```python
stroke_data = pd.read_csv('/content/healthcare-dataset-stroke-data.csv')
stroke_data = stroke_data.drop('id', axis =1)
stroke_data.columns = stroke_data.columns.str.upper() # convert the columns to uppercase

stroke_data
```

```python
x_data = stroke_data.drop('STROKE', axis=1)
y_data = stroke_data['STROKE']

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data,test_size=.2,random_state =123)
# y_data
```

```python

classifier_model = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None, predictions=True, classifiers='all' )
models,predictions = classifier_model.fit(X_train, X_test, y_train, y_test)
# model_dictionary = classifier_model.provide_models(X_train,X_test,y_train,y_test)

models
```

## Regression Problem : LazyRegressor

```python
housing_data = pd.read_csv('/content/Housing.csv')
housing_data.columns = housing_data.columns.str.upper() # convert the columns to uppercase

housing_data
```

```python
# dropping the target from input features
x_data = housing_data.drop('PRICE', axis=1)
# assigning the target
y_data = housing_data['PRICE']

X_train, X_test, y_train, y_test = train_test_split(x_data, y_data,test_size=.2,random_state =123)
```

```python
# define the model and run
lzy_regressor = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None, predictions=False, regressors ='all' )
regressor_model ,predictions = lzy_regressor.fit(X_train, X_test, y_train, y_test)

# model_dictionary = classifier_model.provide_models(X_train,X_test,y_train,y_test)
regressor_model
```
